```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
data=pd.read_csv("/content/diabetes.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Outcome'].value_counts()
```

```python
X=data.drop(['Outcome'],axis=1)
Y=data['Outcome']
```

```python
model=SVC()
```

```python
parameter={
    'kernel':['Linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
classifier=GridSearchCV(model,parameter,cv=5)
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
bestParameter=classifier.best_params_
accuracy=classifier.best_score_

print(bestParameter,accuracy)
```

```python
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
result=result[['param_C','param_kernel','mean_test_score']]
```

```python
result.head()
```

**RandomizedSearchCV**

```python
model=SVC()
```

```python
parameters = {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
classifier2=RandomizedSearchCV(model,parameters,cv=5)
```

```python
classifier2.fit(X,Y)
```

```python
result=pd.DataFrame(classifier2.cv_results_)
```

```python
result=result[['param_C','param_kernel','mean_test_score']]
```

```python
result
```

```python
highParam=classifier2.best_params_
print(highParam)
highAccuracy=classifier2.best_score_
print(highAccuracy)
```
